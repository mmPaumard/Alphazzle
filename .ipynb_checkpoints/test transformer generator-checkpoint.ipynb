```python
from lib.puzzles_generator_MET import prepare_data_p, prepare_data_v
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_p = prepare_data_p('datasets/toy-bas-reliefs/', puzzle_size=256, fragment_per_side=4, fragment_size=46, space=24, batch_size=1)
```

```python
img, label = data_p[0]
img = img.transpose(0,2,3,1).squeeze()
img = (img - np.min(img))/np.ptp(img)
print(label)
plt.imshow(img)
```
